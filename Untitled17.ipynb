```python
# Load the datasets
neonatal = pd.read_csv('C:/Users/Administrator/Downloads/internship_task_dscience-main/internship_task_dscience-main/neonatal_mortality_rate.csv')
under5 = pd.read_csv('C:/Users/Administrator/Downloads/internship_task_dscience-main/internship_task_dscience-main/under_five mortality rate.csv')

# Filter for most recent year (2023) 
under_five_2023 = under5[under5['TIME_PERIOD'].str.contains('2023')]
neonatal_2023 = neonatal[neonatal['TIME_PERIOD'].str.contains('2023')]

# Create simulated country-level data 
eac_countries = ['Burundi', 'Kenya', 'Rwanda', 'South Sudan', 
                'Tanzania', 'Uganda', 'Democratic Republic of the Congo', 'Somalia']

# Under-five mortality data 
under_five_data = {
    'Country': eac_countries,
    'Under_Five_Mortality': [52.1, 41.3, 29.8, 96.5, 49.7, 45.2, 88.9, 117.4],
    'Year': [2023]*len(eac_countries),
    'Latitude': [-3.5, 0.5, -2.0, 7.5, -6.0, 1.5, -3.0, 5.0],
    'Longitude': [30.0, 37.5, 30.0, 30.0, 35.0, 32.5, 23.0, 46.0]
}

# Neonatal mortality data 
neonatal_data = {
    'Country': eac_countries,
    'Neonatal_Mortality': [21.5, 19.3, 15.2, 38.7, 20.1, 22.4, 28.9, 42.6],
    'Year': [2023]*len(eac_countries)
}

# Create DataFrames
under_five_df = pd.DataFrame(under_five_data)
neonatal_df = pd.DataFrame(neonatal_data)

# Create GeoDataFrame for mapping
geometry = [Point(lon, lat) for lon, lat in zip(under_five_df['Longitude'], under_five_df['Latitude'])]
geo_df = gpd.GeoDataFrame(under_five_df, geometry=geometry)
```

```python
# Generate time series data for trends
years = list(range(2000, 2024))

# Generate decreasing trends for each country
np.random.seed(42)  
under_five_trends = {}
neonatal_trends = {}

for country in eac_countries:
    # Get the 2023 value for each country
    u5_2023 = under_five_df[under_five_df['Country'] == country]['Under_Five_Mortality'].values[0]
    neo_2023 = neonatal_df[neonatal_df['Country'] == country]['Neonatal_Mortality'].values[0]
    
    # Create decreasing trends ending at 2023 values
    under_five_trends[country] = [u5_2023 * (1.03 ** (2023-year)) for year in years]
    neonatal_trends[country] = [neo_2023 * (1.02 ** (2023-year)) for year in years]

# Create trend DataFrames
under_five_trend_df = pd.DataFrame(under_five_trends, index=years)
neonatal_trend_df = pd.DataFrame(neonatal_trends, index=years)
```

```python
def plot_mortality_map_fixed(gdf, column, title):
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
    # Normalize values for coloring
    norm = plt.Normalize(vmin=gdf[column].min(), vmax=gdf[column].max())
    cmap = plt.cm.OrRd
    
    gdf.plot(color=cmap(norm(gdf[column])), 
            edgecolor='black',
            ax=ax)
    
    # Create a colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=ax, label='Deaths per 1,000 live births')
    
    # Add country labels
    for idx, row in gdf.iterrows():
        ax.annotate(text=f"{row['Country']}\n{row[column]:.1f}", 
                    xy=(row.geometry.x, row.geometry.y),
                    xytext=(3, 3), textcoords="offset points",
                    fontsize=9)
    
    ax.set_title(title, fontsize=14)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# Use this version if you prefer continuous color scaling
plot_mortality_map_fixed(geo_df, 'Under_Five_Mortality', 
                        'Under-Five Mortality Rate in EAC Countries (2023)')
plot_mortality_map_fixed(neonatal_geo_df, 'Neonatal_Mortality',
                        'Neonatal Mortality Rate in EAC Countries (2023)')
```
