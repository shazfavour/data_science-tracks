```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

# Load the dataset 
file_path = r"C:\Users\Administrator\Downloads\internship_task_dscience-main\internship_task_dscience-main\HIV data 2000-2023.csv"
hiv_df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Extract numeric estimates from the Value column
hiv_df["HIV_Estimate"] = hiv_df["Value"].str.extract(r"^([\d\s,]+)")
hiv_df["HIV_Estimate"] = hiv_df["HIV_Estimate"].replace(r"\s+", "", regex=True).astype(float)

# Drop missing estimates
hiv_df_clean = hiv_df.dropna(subset=["HIV_Estimate"])
```

```python
# Total HIV cases per country (sum across all years)
country_totals = hiv_df_clean.groupby("Location")["HIV_Estimate"].sum().sort_values(ascending=False)

# Cumulative percentage contribution
global_total = country_totals.sum()
cumulative_percent = country_totals.cumsum() / global_total

# Countries contributing up to 75% of global HIV burden
top_75_countries = cumulative_percent[cumulative_percent <= 0.75].index.tolist()
```

```python
# Filter data 
top_df = hiv_df_clean[hiv_df_clean["Location"].isin(top_75_countries)]

# Plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_df, x="Period", y="HIV_Estimate", hue="Location", estimator="sum", errorbar=None)
plt.title("HIV Trends in Countries Contributing to 75% of Global Burden")
plt.xlabel("Year")
plt.ylabel("Estimated People Living with HIV")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
regional_top_df = pd.DataFrame()

# Group by region
for region, group in hiv_df_clean.groupby("ParentLocationCode"):
    region_totals = group.groupby("Location")["HIV_Estimate"].sum().sort_values(ascending=False)
    regional_total = region_totals.sum()
    regional_cum_pct = region_totals.cumsum() / regional_total
    top_countries = regional_cum_pct[regional_cum_pct <= 0.75].index.tolist()
    regional_top_df = pd.concat([regional_top_df, group[group["Location"].isin(top_countries)]])

# Plot trends by WHO region
g = sns.FacetGrid(regional_top_df, col="ParentLocationCode", col_wrap=3, height=4, aspect=1.5)
g.map_dataframe(sns.lineplot, x="Period", y="HIV_Estimate", hue="Location", estimator="sum", errorbar=None)
g.set_axis_labels("Year", "People Living with HIV")
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle("HIV Trends by WHO Region (Top 75% Burden Countries)")
plt.show()
```
